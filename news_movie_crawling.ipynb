```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://search.naver.com/search.naver?where=news&sm=tab_jum&query=%EC%BD%94%EB%A1%9C%EB%82%98"
res = requests.get(url)
html = res.text
soup = BeautifulSoup(html, 'html.parser')
for p in soup.find_all('a',class_='news_tit'):
    print(p.text)
    print(p.get('href'))
```

```python
url = 'https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&query=%EB%B0%95%EC%8A%A4%EC%98%A4%ED%94%BC%EC%8A%A4'
res = requests.get(url)
html = res.text

soup = BeautifulSoup(html, 'html.parser')
title = soup.find_all('div', class_='title_box')

movie_list =[]

for strong in title:
    name = strong.find('strong').text
    num = strong.find('span').text
    movie ={'name':name, 'num':num}
    movie_list.append(movie)
count =0;
for item in movie_list:
    print(item['name'] +" : "+item['num']+"\n")
    count += 1
    if(count == 10) :
        break
```

```python
url = 'https://www.melon.com/chart/'
header = {'User-Agent' : 'Mozilla/5.0'} #멜론은 헤더가 필요함 
res = requests.get(url, headers=header)
html = res.text

soup = BeautifulSoup(html, 'html.parser')

title_list = soup.find_all('div', 'ellipsis rank01')
singer_list = soup.find_all('div', 'ellipsis rank02')

for i in range(100):
    title = title_list[i].find('a').text
    singer = singer_list[i].find('a').text
    print("%3d위: %s - %s \n"%(i+1, title, singer ))
```
